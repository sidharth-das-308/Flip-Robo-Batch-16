# 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
def book_page(url):
    book_names=[]
    author_names=[]
    genres=[]
    reviews=[]
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    books=soup.find_all('h4',class_='italic')
    for i in books:
        book_names.append(i.find('a').text)
    author=soup.find_all('p',class_='sans bold')
    for i in author:
        author_names.append(i.text.replace('\n',''))
    genre=soup.find_all('p',class_='genre-links hidden-phone')
    for i in genre:
        genres.append(i.find('a').text)
    review=soup.find_all('p',class_='excerpt')
    for i in review:
        reviews.append(i.text.replace('\n',''))
    import pandas as pd
    books=pd.DataFrame({})
    books['book_names']=book_names
    books['author_names']=author_names
    books['genres']=genres
    books['reviews']=reviews
    books=books.head()
    return(books)
```

```python
df=book_page('https://bookpage.com/reviews')
df
```

# 7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
def amazon(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    
    products=[]
    product=soup.find_all('span',class_='a-size-medium a-color-base a-text-normal')
    for i in product:
        products.append(i.text)
        
    prices=[]
    price=soup.find_all('span',class_='a-price-whole')
    for i in price:
        prices.append(i.text)
        
    images=[]
    image=soup.find_all('div',class_='a-section aok-relative s-image-fixed-height')
    for i in image:
        images.append(i.find('img'))
    ratings=[]
    avg_ratings=[]
    rating=soup.find_all('span',class_='a-icon-alt')
    for i in rating:
        ratings.append(i.text)
    # Need to skip first four ratings as they belong to something else
    ratings=ratings[4:]
    # Extracting first 3 characters of rating
    for j in ratings:
        avg_ratings.append(j[0:3])
       
    import pandas as pd
    amazon_phones=pd.DataFrame({})
    amazon_phones['products']=products
    amazon_phones['prices']=prices
    amazon_phones['images']=images
    amazon_phones['avg_ratings']=avg_ratings
    return(amazon_phones)
```

```python
# Phones under 20000
df=amazon('https://www.amazon.in/Mobile-Phone-Under-20000-Rupees/s?k=Mobile+Phone+Under+20000+Rupees')
df
```

# 8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
page = requests.get('https://forecast.weather.gov/MapClick.php?lat=37.77493000000004&lon=-122.41941999999995#.YJZJp7UzbIU')
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
ind = [1,3,5,7]
period = []
short_desc = []
desc = []
temp = []

s = soup.find_all('p', class_='period-name')
for info in s:
    name = info.text
    period.append(name)

t = soup.find_all('p', class_='short-desc')
for info in t:
    name = info.text
    short_desc.append(name)

q = soup.find_all('p', class_='temp temp-low')
r = soup.find_all('p', class_='temp temp-high')
for info in r:
    name = info.text
    temp.append(name)

for info in q:
    name = info.text
    temp.insert(ind[0], name)
    del ind[0]


u = soup.find_all('div', class_='col-sm-10 forecast-text')
for info in u:
    name = info.text
    desc.append(name)
```

```python
Desc = desc[0:9]
```

```python
Weather = pd.DataFrame({})
Weather['Period'] = period
Weather['Short Description'] = short_desc
Weather['Temperature'] = temp
Weather['Description'] = Desc
Weather
```
